### Using Beautiful Soup

### .prettify()

```python
#NOTE: Prettify only works for the full document and the .find() method
print(soup.prettify())
```

### Converting to a List

```python
# Tags may contain strings and other tags. These elements are the tag’s children.
list(soup.children)

print(len(list(soup.children)))
```

### Extracting Beautiful Soup Elements

```python
# We can traverse through an HTML page and extract other tags and text
# The below example shows the types of iterables available in the object created from the HTML Document
# .Tag allows us to dive deeper into the document i.e we can look for HTML attributes like .class and if needed go deeper into the document from there
for child in soup.children:
    print(type(child))
```

### Assinging Variables from Beautiful Soup

```python
# import pprint

html = list(soup.children)[2] # selecting the HTML element child from the soup object
body = list(html.children)[3] # selecting the body element child from the soup object
center = list(body.children)[4]
table = list(center.children)[0]

print(table.prettify())
```

### .find() <br>
<p>Find a specific instance of the parameter passed in</p>

```python
table.find('b')
```

```python
table.find(size='3')
```

### .find_all() <br>
<p>Similar to .find(), except this will return all of them instead of one</p>

```python
table.find_all('b')
```

```python
text_body = []

for b in table.find_all('b'):
    text_body.append(b.text.replace('\xa0\xa0\xa0\xa0', ''))
    
text_body
```

## Exercise <br>
<p>Using the Beautiful Soup library, grab the data from the following link: https://www.nbastuffer.com/2019-2020-nba-player-stats/. After getting the data, display the players name and team inside of a pandas dataframe.</p>

```python
# Hint: Use the .get_text() method

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

```

```python
nba = httpx.get('https://www.nbastuffer.com/2020-2021-nba-player-stats/')
nba
```

```python
nba_soup = BeautifulSoup(nba.content, 'html.parser')
print(nba_soup.prettify())
```

```python
html = list(nba_soup.children)[3]
table = html.find('body').find('table')
```

```python
names = []
positions = []
for table_row in table.find_all('tr')[1:]:
    names.append(table_row.find_all(text=True)[1])
    positions.append(table_row.find_all(text=True)[3])
```

```python
my_dict = {
    'name': names,
    'position': positions
}

df = pd.DataFrame.from_dict(my_dict)
df
```

```python
df.groupby('position').count()
```

```python
from selenium import webdriver
from time import sleep

# Create an instance of Chrome Webdriver using the webdriver in the local directory
driver = webdriver.Chrome('./chromedriver.exe')
# Open up the Flask Blog Login Page
driver.get('https://kekambas-72-bs.herokuapp.com/auth/login')

sleep(5)
# Find the username input
username = driver.find_element_by_name('username')
# Clear the username input
username.clear()
# Add 'bottest' to the username input
username.send_keys('bottest')

# Same as username but with the password field
password = driver.find_element_by_name('password')
password.clear()
password.send_keys('abc123')
# password.send_keys(Keys.RETURN)

# Find the submit button
submit_btn = driver.find_element(By.ID, 'submit')
# Click the submit button
submit_btn.click()


# Navigate to the creat a post page
driver.get('https://kekambas-72-bs.herokuapp.com/blog/createpost')

# Get title input and update with new title
title = driver.find_element_by_name('title')
title.clear()
title.send_keys('This test was written by a bot')

# Same with content
content = driver.find_element_by_name('content')
content.clear()
content.send_keys('Beep boop boop beep. I am a robot here to take over the world. Meep Morp')

content.send_keys(Keys.RETURN)

sleep(10)

driver.close()
```
